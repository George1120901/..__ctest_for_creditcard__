```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Microsoft JhengHei'

# 讀取CSV檔案
df = pd.read_csv('sex.csv',encoding='utf-8')
print (df.head(5))

# 以地區分組，計算每個地區的總交易金額
grouped_by_region = df.groupby('年月')['信用卡交易金額[新台幣]'].sum().reset_index()

# 繪製地區總交易金額的圖表
plt.figure(figsize=(10, 6))
plt.plot(grouped_by_region['年月'], grouped_by_region['信用卡交易金額[新台幣]'])
plt.title('總交易金額')
plt.xlabel('年月')
plt.ylabel('信用卡交易金額[新台幣]')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Microsoft JhengHei'
# 讀取CSV檔案
df = pd.read_csv('sex.csv', encoding='utf-8')


# 將 '年月' 欄位轉換為日期型態


# 以年月分組，計算每個年月的總信用卡交易金額
grouped_by_month = df.groupby(['年月']).agg(
    {'信用卡交易金額[新台幣]': 'sum'}).reset_index()

# 繪製總信用卡交易金額的線形圖
plt.figure(figsize=(10, 6))
plt.plot(grouped_by_month['年月'], grouped_by_month['信用卡交易金額[新台幣]'], marker='o')
plt.title('總信用卡交易金額')
plt.xlabel('年月')
plt.ylabel('信用卡交易金額[新台幣]')
plt.xticks(rotation=45)  # 若日期標籤太擁擠，可進行旋轉
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.family'] = 'Microsoft JhengHei'

# 資料
categories = ['100-125萬', '125-150萬', '150-175萬', '175-200萬', '200萬以上', '50-75萬', '75-100萬', '未達50萬']
total_amount = [4.32622E+11, 2.18232E+11, 1.9481E+11, 1.23225E+11, 7.82573E+11, 1.4403E+12, 6.54052E+11, 2.87012E+12]
average_per_transaction = [1913.48306, 2102.250922, 2525.274395, 2687.741216, 4478.122163, 1422.572996, 1600.217482, 1222.39183]

# 設定圖表
fig, ax1 = plt.subplots()

# 畫長條圖
bar_width = 0.35
bar1 = ax1.bar(np.arange(len(categories)), total_amount, bar_width, label='信用卡交易金額', color='b')

# 設定 X 軸刻度
ax1.set_xticks(np.arange(len(categories)))
ax1.set_xticklabels(categories, rotation=45, ha='right')

# 設定 Y 軸標籤
ax1.set_ylabel('信用卡交易金額 [新台幣]', color='b')
ax1.tick_params('y', colors='b')

# 建立第二個 Y 軸
ax2 = ax1.twinx()
ax2.plot(np.arange(len(categories)), average_per_transaction, 'r--o', label='平均每筆金額')

# 設定 Y 軸標籤
ax2.set_ylabel('平均每筆金額', color='r')
ax2.tick_params('y', colors='r')

# 添加圖例
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# 顯示圖表
plt.title('信用卡交易金額與平均每筆金額')
plt.show()
```
