```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# 設定中文字型
plt.rc('font', family='Microsoft JhengHei')

# 連結資料庫
conn = sqlite3.connect('creditcard.db') 

query_condition = "SELECT 年齡層 ,信用卡金額 FROM age group by 年齡層"
df = pd.read_sql_query(query_condition, conn)
conn.close()


# 繪製圓餅圖
plt.pie(df['信用卡金額'], labels=df['年齡層'], autopct='%1.1f%%',
            wedgeprops=dict(width=size, edgecolor='w'))
plt.title('年齡層分佈')
plt.show()

plt.pie(df['信用卡金額'], labels=df['年齡層'],autopct='%1.1f%%')
plt.title('年齡層分佈')
plt.show()

# 繪製折線圖
plt.plot(df['信用卡金額'],  marker='o', linestyle='-', color='b')
plt.title('年齡層分佈')
plt.show()

plt.bar(df['年齡層'],df['信用卡金額'],  color='b', width=0.4, align='edge')
plt.show()
```
